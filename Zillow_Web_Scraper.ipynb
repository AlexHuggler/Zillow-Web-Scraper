### Preprocessing to retrieve latitude and longitude values. It's worth noting that negative latitudes represent the southern hemisphere, and negative longitudes represent the western hemisphere, so some adapatation of the below regex will be needed if viewing data in Asia or Australia

```python
# Prep Lat/long columns
df['latLong'][0]
```

```python
df[['latitude','longitude']] = df['latLong'].str.split(',',expand=True)
df['latitude']=df['latitude'].astype('string')
df['longitude']=df['longitude'].astype('string')
```

```python
df['longitude'][0][14:23]
```

```python
pattern=r'([-\d.+])'

df['longitude']=df['longitude'].apply(lambda x : re.findall(pattern, x)).astype(str)
```
